```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

print("Academic Integrity Checker - AI vs Human Text Detection")

# Load dataset
data = pd.read_csv("ai_human.csv")

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    data['text'], data['label'], test_size=0.25, random_state=42
)

# TF-IDF
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Train model
model = LogisticRegression()
model.fit(X_train_vec, y_train)

# Accuracy
y_pred = model.predict(X_test_vec)
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", round(accuracy, 2))

# Test
text = ["AI is changing education"]
vec = vectorizer.transform(text)
pred = model.predict(vec)

if pred[0] == 1:
    print("Result: AI Generated Text")
else:
    print("Result: Human Written Text")
```

```python
import pandas as pd
import gradio as gr
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

print("Academic Integrity Checker - AI vs Human Text Detection\n")

# Load dataset
data = pd.read_csv("ai_human.csv")
print("Dataset Loaded Successfully!")
print("Total Data:", len(data))
print(data.head(), "\n")

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    data['text'], data['label'], test_size=0.25, random_state=42
)

print("Training Data Size:", len(X_train))
print("Testing Data Size:", len(X_test))

# TF-IDF
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

print("Text converted to numerical features using TF-IDF")

# Train model
model = LogisticRegression()
model.fit(X_train_vec, y_train)

print("Model Training Completed!")

# Accuracy
y_pred = model.predict(X_test_vec)
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", round(accuracy, 2))

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Prediction function
def detect_text(text):
    vec = vectorizer.transform([text])
    pred = model.predict(vec)

    if pred[0] == 1:
        return "AI Generated Text"
    else:
        return "Human Written Text"

# UI
interface = gr.Interface(
    fn=detect_text,
    inputs="text",
    outputs="text",
    title="Academic Integrity Checker",
    description="AI vs Human Text Detection System using Machine Learning"
)

interface.launch()
```
